```python
import pandas as pd
import json
import copy
```

```python
df = pd.read_csv("20182019_prediction.csv")
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df2 = copy.deepcopy(df)
keys = df2.keys()
df2
```

```python
keys = list(keys)
keys.remove('PLAYER')
keys.remove('TEAM')
keys.remove('PLAYOFF')
keys.remove('SEMI-FINAL')
print(keys)
```

```python
df2[keys] = scaler.fit_transform(df2[keys])
df2
```

```python
df
```

```python
team_player1 = {}
for i, player in df.iterrows():
    if player.TEAM not in team_player1:
        team_player1[player.TEAM] = []
    team_player1[player.TEAM].append(player.to_dict())
```

```python
team_player2 = {}
for i, player in df2.iterrows():
    if player.TEAM not in team_player2:
        team_player2[player.TEAM] = []
    team_player2[player.TEAM].append(player.to_dict())
```

```python
team_player1
```

```python
team_player2
```

```python
with open('20182019_prediction.json', 'w') as f:
    json.dump(team_player1, f, indent=4)
```

```python
with open('20182019_prediction_scaled.json', 'w') as f:
    json.dump(team_player2, f, indent=4)
```
