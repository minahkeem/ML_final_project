```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

# We are creating a predictive model for NBA playoffs 2018-2019
We will be using the player data for the 2016-17 season to train our models and 2017-18 data to test our models.

```python
df_tr = pd.read_csv("20162017_data.csv")
df_ts = pd.read_csv("20172018_data.csv")
```

```python
df_tr
```

```python
one_hot_team_tr = pd.get_dummies(df_tr['TEAM'])
one_hot_team_ts = pd.get_dummies(df_ts['TEAM'])
one_hot_team_tr
```

```python
new_df_tr=df_tr.drop(labels=['TEAM', 'PLAYER'], axis=1)
new_df_tr=one_hot_team_tr.join(new_df_tr)
new_df_ts = df_ts.drop(labels=['TEAM', 'PLAYER'], axis=1)
new_df_ts = one_hot_team_ts.join(new_df_ts)
```

```python
new_df_ts
```

```python
col_names = list(new_df_tr)
print(col_names)
```

```python
scaler = StandardScaler()

#this only normalizes three categories: games played, wins, losses

#new_df_tr[['GAMES PLAYED', 'WINS', 'LOSSES']]=scaler.fit_transform(new_df_tr[['GAMES PLAYED', 'WINS', 'LOSSES']])
#new_df_ts[['GAMES PLAYED', 'WINS', 'LOSSES']]=scaler.fit_transform(new_df_ts[['GAMES PLAYED', 'WINS', 'LOSSES']])

#normalizing all features
data_tr = new_df_tr.values
data_tr = np.hstack((scaler.fit_transform(data_tr[:,:data_tr.shape[1]-2]),data_tr[:,data_tr.shape[1]-2:]))
data_ts = new_df_ts.values
data_ts = np.hstack((scaler.fit_transform(data_ts[:,:data_ts.shape[1]-2]),data_ts[:,data_ts.shape[1]-2:]))
```

```python
data_tr.shape, data_ts.shape
```

```python
data_tr
```

```python
#dividing data into X and y
X_tr = np.array(data_tr[:,:data_tr.shape[1]-2])
y_tr = np.array(data_tr[:,data_tr.shape[1]-2:])
X_ts = np.array(data_ts[:,:data_ts.shape[1]-2])
y_ts = np.array(data_ts[:,data_ts.shape[1]-2:])
X_tr.shape, y_tr.shape, X_ts.shape, y_ts.shape
```

```python
#Model 1: Simple Logistic Regression Model (binary class, two different y values for playoffs and semi-finals)
from sklearn.linear_model import LogisticRegressionCV

#using L1 regularization
model_playoff = LogisticRegressionCV(penalty='l1',solver='liblinear')
model_semi = LogisticRegressionCV(penalty='l1',solver='liblinear')
```

```python
y_playoff = y_tr[:,0]
y_semi = y_tr[:,1]
```

```python
model_playoff.fit(X_tr,y_playoff)
model_semi.fit(X_tr,y_semi)
```

```python
yhat_playoff = model_playoff.predict(X_ts)
model_playoff.score(X_ts, y_ts[:,0])
```

```python
yhat_semi = model_semi.predict(X_ts)
model_semi.score(X_ts, y_ts[:,1])
```

```python
model_playoff.coef_
```

# The above coefficients after LASSO regularization shows that the actual player stats are wiped out and only the one-hot-encoded team names are taken into account for predicting the playoffs and semi-finals.

We will redo the fitting with a NEW data set that excludes the team name altogether and focus only on individual players' stats.

```python
df_tr = pd.read_csv("20162017_data.csv")
df_ts = pd.read_csv("20172018_data.csv")
```

```python
new_df_tr = df_tr.drop(labels=['TEAM', 'PLAYER'], axis=1)
new_df_ts = df_ts.drop(labels=['TEAM', 'PLAYER'], axis=1)
```

```python
new_df_ts
```

```python
col_names = list(new_df_tr)
print(col_names)
```

```python
scaler = StandardScaler()

#this only normalizes three categories: games played, wins, losses

#new_df_tr[['GAMES PLAYED', 'WINS', 'LOSSES']]=scaler.fit_transform(new_df_tr[['GAMES PLAYED', 'WINS', 'LOSSES']])
#new_df_ts[['GAMES PLAYED', 'WINS', 'LOSSES']]=scaler.fit_transform(new_df_ts[['GAMES PLAYED', 'WINS', 'LOSSES']])

#normalizing all features
data_tr = new_df_tr.values
data_tr = np.hstack((scaler.fit_transform(data_tr[:,:data_tr.shape[1]-2]),data_tr[:,data_tr.shape[1]-2:]))
data_ts = new_df_ts.values
data_ts = np.hstack((scaler.fit_transform(data_ts[:,:data_ts.shape[1]-2]),data_ts[:,data_ts.shape[1]-2:]))
```

```python
data_tr.shape, data_ts.shape
```

```python
data_tr
```

```python
#dividing data into X and y
X_tr = np.array(data_tr[:,:data_tr.shape[1]-2])
y_tr = np.array(data_tr[:,data_tr.shape[1]-2:])
X_ts = np.array(data_ts[:,:data_ts.shape[1]-2])
y_ts = np.array(data_ts[:,data_ts.shape[1]-2:])
X_tr.shape, y_tr.shape, X_ts.shape, y_ts.shape
```

```python
#Model 1: Simple Logistic Regression Model (binary class, two different y values for playoffs and semi-finals)
from sklearn.linear_model import LogisticRegressionCV

#using Ridge regularization
model_playoff_ridge = LogisticRegressionCV(penalty='l2',solver='liblinear')
model_semi_ridge = LogisticRegressionCV(penalty='l2',solver='liblinear')
```

```python
y_playoff = y_tr[:,0]
y_semi = y_tr[:,1]
```

```python
model_playoff_ridge.fit(X_tr,y_playoff)
model_semi_ridge.fit(X_tr,y_semi)
```

```python
yhat_playoff = model_playoff_ridge.predict(X_ts)
model_playoff_ridge.score(X_ts, y_ts[:,0])
```

```python
yhat_semi = model_semi_ridge.predict(X_ts)
model_semi_ridge.score(X_ts, y_ts[:,1])
```

```python
model_playoff_ridge.coef_
```

```python
from sklearn.metrics import confusion_matrix as cm

playoff_matrix = cm(y_ts[:,0], yhat_playoff)
semi_matrix = cm(y_ts[:,1], yhat_semi)
```

```python
playoff_matrix
```

```python
semi_matrix
```

```python
#using LASSO regularization
model_playoff_lasso = LogisticRegressionCV(penalty='l1',solver='liblinear')
model_semi_lasso = LogisticRegressionCV(penalty='l1',solver='liblinear')
```

```python
model_playoff_lasso.fit(X_tr,y_playoff)
model_semi_lasso.fit(X_tr,y_semi)
```

```python
yhat_playoff = model_playoff_lasso.predict(X_ts)
model_playoff_lasso.score(X_ts, y_ts[:,0])
```

```python
yhat_semi = model_semi_lasso.predict(X_ts)
model_semi_lasso.score(X_ts, y_ts[:,1])
```

```python
model_playoff_lasso.coef_
```

```python
playoff_matrix = cm(y_ts[:,0], yhat_playoff)
semi_matrix = cm(y_ts[:,1], yhat_semi)
```

# It seems like logistic regression gives better accuracy scores than SVM, so we will make predictions on the current 2018-2019 season using logistic regression.

For playoffs, <b>LASSO regularization</b> makes more sense to use (according to both accuracy score and confusion matrix) and for semi-finals, <b>Ridge regularization</b> makes more sense to use.

Things we've tried in this model:
- one hot encoding using LASSO regularization--but it turned out to be a disaster since most features <i>besides</i> the features derived from one hot encoding were wiped out (which makes sense--but it is not what we want to see)
- Ridge Regularization on the data set that excludes the name of player and the team name
- LASSO Regularization on the data set that excludes the name of player and the team name
