```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

# We are creating a predictive model for NBA playoffs 2018-2019
We will be using the player data for the 2016-17 season to train our models and 2017-18 data to test our models.

```python
df_tr = pd.read_csv("20162017_data.csv")
df_ts = pd.read_csv("20172018_data.csv")
```

```python
df_tr
```

```python
one_hot_team_tr = pd.get_dummies(df_tr['TEAM'])
one_hot_team_ts = pd.get_dummies(df_ts['TEAM'])
one_hot_team_tr
```

```python
new_df_tr=df_tr.drop(labels=['TEAM', 'PLAYER'], axis=1)
new_df_tr=one_hot_team_tr.join(new_df_tr)
new_df_ts = df_ts.drop(labels=['TEAM', 'PLAYER'], axis=1)
new_df_ts = one_hot_team_ts.join(new_df_ts)
```

```python
new_df_ts
```

```python
col_names = list(new_df_tr)
print(col_names)
```

```python
scaler = StandardScaler()
new_df_tr[['GAMES PLAYED', 'WINS', 'LOSSES']]=scaler.fit_transform(new_df_tr[['GAMES PLAYED', 'WINS', 'LOSSES']])
new_df_ts[['GAMES PLAYED', 'WINS', 'LOSSES']]=scaler.fit_transform(new_df_ts[['GAMES PLAYED', 'WINS', 'LOSSES']])
```

```python
new_df_tr['GAMES PLAYED']
```

```python
data_tr = new_df_tr.values
data_ts = new_df_ts.values
data_tr.shape, data_ts.shape
```

```python
#dividing data into X and y
X_tr = np.array(data_tr[:,:data_tr.shape[1]-2])
y_tr = np.array(data_tr[:,data_tr.shape[1]-2:])
X_ts = np.array(data_ts[:,:data_ts.shape[1]-2])
y_ts = np.array(data_ts[:,data_ts.shape[1]-2:])
X_tr.shape, y_tr.shape, X_ts.shape, y_ts.shape
```

```python
#Model 1: Simple Logistic Regression Model (binary class, two different y values for playoffs and semi-finals)
from sklearn.linear_model import LogisticRegressionCV

#using L2 regulari
model_playoff = LogisticRegressionCV(penalty='l2')
model_semi = LogisticRegressionCV(penalty='l2')
```

```python
y_playoff = y_tr[:,0]
y_semi = y_tr[:,1]
```

```python
model_playoff.fit(X_tr,y_playoff)
model_semi.fit(X_tr,y_semi)
```

```python
yhat_playoff = model_playoff.predict(X_ts)
model_playoff.score(X_ts, y_ts[:,0])
```

```python
yhat_semi = model_semi.predict(X_ts)
model_semi.score(X_ts, y_ts[:,1])
```

```python
model_playoff.coef_
```

```python
from sklearn.metrics import precision_recall_fscore_support

#metrics for playoffs
prec , recal , fscore,_ = precision_recall_fscore_support(y_ts[:,0], yhat_playoff, average='binary')
prec, recal, fscore
```

```python
#metrics for semi-finals
prec , recal , fscore,_ = precision_recall_fscore_support(y_ts[:,1], yhat_semi, average='binary')
prec, recal, fscore
```
